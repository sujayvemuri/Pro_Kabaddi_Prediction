```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import re
import time
```

Download the Chrome Webdriver from https://sites.google.com/a/chromium.org/chromedriver/downloads (make sure you download the same Chrome version driver)

And place the file at notebook location

```python
driver = webdriver.Chrome(executable_path='chromedriver.exe')
```

```python
driver.get('https://www.prokabaddi.com/standings')
```

```python
time.sleep(5)
driver.find_element_by_xpath('//*[@id="2e6c791f-e102-49e7-9fe8-0bfb98aac1cd"]/component/div/div/div/div/div/div[1]/div/div[1]/ul/li').click()
time.sleep(1)
driver.find_element_by_xpath('//*[@id="season_selectBox_web"]/span[7]').click()
time.sleep(1)
```

Wait for above cell to execute

```python
soup = BeautifulSoup(driver.page_source)
standings = soup.findAll('div',attrs={'class':'sipk-standing-dataRow common-standing-dataRow si-formguide-active'})
```

```python
rank = []
team = []
play = []
won = []
lost = []
draw = []
scorediff = []
point = []

for item in standings:
    rank.append(item.find('div',attrs={'class':'sipk-table-col sipk-rank'}).span.text)
    team.append(item.find('div',attrs={'class':'sipk-table-col sipk-team'}).span.text)
    play.append(item.find('div',attrs={'class':'sipk-table-col sipk-play'}).span.text)
    won.append(item.find('div',attrs={'class':'sipk-table-col sipk-won'}).span.text)
    lost.append(item.find('div',attrs={'class':'sipk-table-col sipk-lost'}).span.text)
    draw.append(item.find('div',attrs={'class':'sipk-table-col sipk-drow'}).span.text)
    scorediff.append(item.find('div',attrs={'class':'sipk-table-col sipk-scoreDiff'}).span.text)
    point.append(item.find('div',attrs={'class':'sipk-table-col sipk-points'}).span.text)
```

```python
standing1 = pd.DataFrame({'season':1, 'rank':rank, 'team':team, 'played':play, 'won':won
                          , 'lost':lost, 'draw':draw, 'scorediff':scorediff, 'points':point})
standing1
```
