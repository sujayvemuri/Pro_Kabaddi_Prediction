#### Task1 - Winner Predictor

```python
term7=pd.read_csv(r"data\season7_team_stats.csv")
standings=pd.read_csv(r"data\standings_overall.csv")
```

```python
term7 = term7.rename(columns={'team_name':'team'})
standings.team.loc[standings['team'] == 'Jaipur Pink Panthers'] = standings['team'].replace('Jaipur Pink Panthers', 'jaipur-pink-panthers', inplace=True)
standings.team.loc[standings['team'] == 'Dabang Delhi K.C.'] = standings['team'].replace('Dabang Delhi K.C.', 'dabang-delhi-kc', inplace=True)
standings.team.loc[standings['team'] == 'Bengal Warriors'] = standings['team'].replace('Bengal Warriors', 'bengal-warriors', inplace=True)
standings.team.loc[standings['team'] == 'Haryana Steelers'] = standings['team'].replace('Haryana Steelers', 'haryana-steelers', inplace=True)
standings.team.loc[standings['team'] == 'U.P. Yoddha'] = standings['team'].replace('U.P. Yoddha', 'up-yoddha', inplace=True)
standings.team.loc[standings['team'] == 'U Mumba'] = standings['team'].replace('U Mumba', 'u-mumba', inplace=True)
standings.team.loc[standings['team'] == 'Gujarat Fortunegiants'] = standings['team'].replace('Gujarat Fortunegiants', 'gujarat-fortunegiants', inplace=True)
standings.team.loc[standings['team'] == 'Puneri Paltan'] = standings['team'].replace('Puneri Paltan', 'puneri-paltan', inplace=True)
standings.team.loc[standings['team'] == 'Bengaluru Bulls'] = standings['team'].replace('Bengaluru Bulls', 'bengaluru-bulls', inplace=True)
standings.team.loc[standings['team'] == 'Patna Pirates'] = standings['team'].replace('Patna Pirates', 'patna-pirates', inplace=True)
standings.team.loc[standings['team'] == 'Telugu Titans'] = standings['team'].replace('Telugu Titans', 'telugu-titans', inplace=True)
standings.team.loc[standings['team'] == 'Tamil Thalaivas'] = standings['team'].replace('Tamil Thalaivas', 'tamil-thalaivas', inplace=True)
stand7 = standings.loc[standings['season']==7]
```

```python
final7 = pd.merge(term7, stand7, on=['team'], how='inner')
final7.head()
```

```python
X_test = final7.drop(['season', 'team'], axis=1)
```

```python
# load the model from disk
task1_model = joblib.load(r'models\task1_model.sav')
```

```python
xgpred1 = task1_model.predict_proba(X_test)[:,1]
```

```python
win1 = pd.DataFrame({'Team':final7.team, 'Win_Pred':xgpred1, 'Win':'No'})
win1['Win'][win1.Win_Pred == max(xgpred1)] = 'Yes'
win1
```

#### Task - 6 Predicting the player with High Success Raid Percentage
    To perform this task, we will first look at the top 20 players with Successful Raids

```python
player7=pd.read_csv(r"data\season7_player_stats.csv")
```

```python
top20_raids = player7.sort_values(by=['success_raids']
                                  , ascending=False)[:20]
top20_raids.head()
```

```python
top20_raids['success_raids_percent'] = round(100*(top20_raids['success_raids']/top20_raids['total_raids']), 2)
top20_raids = top20_raids.sort_values(by=['success_raids_percent'], ascending=False)
```

```python
top20_raids.head()
```

Lets plot the Success Raids percent on a Bar graph

```python
sns.set(rc={'figure.figsize':(13,10)})
sns.barplot(y='name', x='success_raids_percent', data=top20_raids, palette='Blues_d')
plt.xlabel("Player Name")
plt.ylabel("Success Raid percent")
plt.title("Top 20 Players with High Raid Percent")
plt.show()
```

From the above graph we can say that **Pavan Kumar Sehrawat** will be the player with High Success Raid Percentage of around **55%**
