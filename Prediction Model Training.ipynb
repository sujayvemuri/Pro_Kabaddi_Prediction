#### Task 5 Predicting the team with max Super Performance Total (SPT)

```python
term1=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season1_team_stats.csv")
term2=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season2_team_stats.csv")
term3=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season3_team_stats.csv")
term4=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season4_team_stats.csv")
term5=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season5_team_stats.csv")
term6=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season6_team_stats.csv")
term7=pd.read_csv("D:\Hackathon\Pro_Kabbadi_League\data\season7_team_stats.csv")
```

```python
SPT = pd.concat([term1,term2,term3,term4,term5,term6], ignore_index=True)
SPT.head()
```

```python
SPT['SPT'] = SPT['team_super_raids']+SPT['team_super_tackles']+SPT['team_allouts_inflicted']-SPT['team_allouts_conceded']
term7['SPT'] = term7['team_super_raids']+term7['team_super_tackles']+term7['team_allouts_inflicted']-term7['team_allouts_conceded']
```
